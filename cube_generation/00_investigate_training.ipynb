```python
import pandas as pd
import IPython
import shapely
import json

import constants
```

```python
sites_params = pd.read_csv(constants.PATH_SITES_PARAMETERS_TRAINING)
sites_params
```

```python
y = sites_params["geographical location: centre coordiantes: \r\nLAT \r\n(EPSG4326/WGS84)"].values
x = sites_params["geographical location: centre coordiantes: \r\nLON \r\n(EPSG4326/WGS84)"].values

points = [
    {"type": "Feature", 
     "geometry": {"type": "Point", "coordinates": [x[i], y[i]]}, 
    } for i in range(len(x))

]

# Create a GeoJSON FeatureCollection
geojson_data = {
    "type": "FeatureCollection",
    "features": points
}
```

```python
IPython.display.GeoJSON(data=json.dumps(geojson_data))
```
